````python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from transformers.generation.utils import GenerationConfig

model_path = "/root/autodl-tmp/models/DeepSeek-R1-Distill-Qwen-7B"
quantization_config = BitsAndBytesConfig(load_in_8bit=True)
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_path, device_map="auto", trust_remote_code=True, quantization_config= quantization_config)
generation_config = GenerationConfig.from_pretrained(model_path, trust_remote_code=True, max_new_tokens = 512) # 可指定不同的生成长度、top_p等相关超参
````

````python
prompt = '''
你是一名高级数据库工程师，请你根据所提供的表结构说明以及用户问题，生成sql语句，SQL语句中的表名和字段名应该满足表结构中的字段名和表名，数据库为sqlite，你生成的sql语句格式必须符合sqlite格式。
------表结构说明开始------
{'school_score': '\\n字段 类型\\n学校 TEXT \\n年份 TEXT\\n录取批次 TEXT\\n招生类型 TEXT\\n最低分 TEXT\\n最低分对应位次 TEXT\\n省控线 TEXT\\n专业组 TEXT\\n选科要求 TEXT\\n', 'professional_score': '\\n字段 类型\\n学校 TEXT\\n专业 TEXT\\n备注 TEXT\\n年份 TEXT\\n科目 TEXT\\n平均分 TEXT\\n最低分 TEXT\\n最低分对应位次 TEXT\\n', '分数位次表': '\\n字段 类型\\n省份 TEXT\\n分数 TEXT\\n位次 TEXT\\n人数 TEXT\\n'}
------表结构说明结束------

必须注意：答案只需要sql语句，不需要其他任何输出。
如：
用户问题：湖南大学2022年录取分数线。
输出：SELECT 学校, 年份, 录取批次, 招生类型,  最低分, 最低分对应位次 FROM school_score WHERE 学校 = '湖南大学' AND 年份 = 2022
请回答：
用户问题：我考了620分可以上哪些学校？。
输出：
'''
messages = [{'role': 'user', 'content': prompt}]
print(messages)
````

````python
text = tokenizer.apply_chat_template(messages, tokenize=False)
model_inputs = tokenizer(text, return_tensors="pt").to(model.device)
outputs = model.generate(**model_inputs, **generation_config.to_dict())
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
````

````python
result = response.split('```')[1].split('\n')[1].strip()
````
